```python
import matplotlib.pyplot as mat
```

```python
x = [0,1,2,3,4,5]
y = [0, 1, 4, 9, 16, 25]
x1 = [0, 10, 20, 30, 40, 50]
y1 = [0, 20, 10, 40, 10, 50]
mat.subplot(1, 2, 1)
mat.plot(x,y, marker='.',mfc="r", mec="k", ms=10)
mat.title("Line Chart of Subplot")
mat.xlabel("X- axis")
mat.ylabel("y-axis")
mat.subplot(1,2,2)
mat.plot(x1, y1, marker='.',mfc='y', mec='r', ms=10)
mat.title("Line Chart of Subplot")
mat.xlabel("X- axis")
mat.ylabel("y-axis")
mat.show()
```
