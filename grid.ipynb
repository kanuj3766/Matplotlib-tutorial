```python
import matplotlib.pyplot as mat
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 2, 3, 4, 5]
mat.plot(x,y, marker=".", mec="r", mfc="k", ms=7, ls="solid",lw=2, c="y")
mat.grid(axis="both", ls="solid", lw=1, c="g")
mat.xlabel("x-axis")
mat.ylabel("y-axis")
mat.title("Grid chart")
mat.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 2, 3, 4, 5]
mat.plot(x,y, marker=".", mec="r", mfc="k", ms=7, ls="solid",lw=2, c="y")
mat.grid(axis="x", ls="dotted", lw=2, c="blue")
mat.xlabel("x-axis")
mat.ylabel("y-axis")
mat.title("Grid chart")
mat.show()
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 2, 3, 4, 5]
mat.plot(x,y, marker="*", mec="g", mfc="b", ms=7, ls="solid",lw=2, c="y")
mat.grid(axis="y", ls="dashdot", lw=1, c="r")
mat.xlabel("x-axis")
mat.ylabel("y-axis")
mat.title("Grid chart")
mat.show()
```
