```python
import matplotlib.pyplot as mat
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 8]
x1 = [0, 1, 2, 3, 4, 5, 6, 8]
y = [0, 5, 10, 15, 20, 25, 30, 35]

mat.plot(x, y, x1)
mat.legend(["product","sales"])
mat.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 8]
x1 = [0, 1, 2, 3, 4, 5, 6, 8]
y = [0, 5, 10, 15, 20, 25, 30, 35]

mat.plot(x,y, label="product")
mat.plot(x,x1, label="sales")
mat.legend()
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 8]
x1 = [0, 1, 2, 3, 4, 5, 6, 8]
y = [0, 5, 10, 15, 20, 25, 30, 35]

mat.plot(x,y, label="product")
mat.plot(x,x1, label="sales")
mat.legend(loc="upper right")
mat.plot(loc=2, facecolor="red", edgecolor='blue', framealpha=1, shadow = True, fancybox= False)
mat.show()
```
