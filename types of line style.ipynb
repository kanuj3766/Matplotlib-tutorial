```python
import matplotlib.pyplot as mat
```

# LINE CHART

```python
x = [0, 1, 2, 3, 4, 5]
y = [i**2 for i in x]
mat.plot(x,y, marker=".", mfc="y", mec="k", ms=10,ls="solid", c="r")
fc= {"family": "Times in roman", "size":20, "color":'r'}
fc1= {"family": "Times in roman", "size":25, "color":'g'}

mat.xlabel("x-axis",fontdict=fc)
mat.ylabel("y-label", fontdict=fc)
mat.title("line-chart", fontdict=fc1)
mat.show()
```
