```python
import matplotlib.pyplot as mat
```

```python
x = [0,1,2,3,4,5]
y = [0,1,4,9,16,25]
y1 = [0,10,20,30,40,50]
y2 = [0,20,10,40,30,50]
mat.plot(x,y,x,y1)
mat.legend(["sales", ["product"]])
mat.plot(x,y,label="product", marker=".",mfc='r',mec='k')
mat.plot(x,y1,label="sales", marker="*",mfc='w',mec='c')
mat.plot(x,y2, label="company",marker="s",mfc='g',mec='y')
mat.legend(loc=2, facecolor="cyan",edgecolor='red', framealpha=0.5,shadow=True)
mat.show()
```
